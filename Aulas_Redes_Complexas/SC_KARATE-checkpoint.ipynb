```python
import networkx as nx
import matplotlib.pyplot as plt
from matplotlib.cm import ScalarMappable

# Leitura o arquivo GML
G = nx.read_gml("karate.gml", label="id")

# Calculo  do grau de cada nÃ³
node_degrees = dict(G.degree())

# DefiniÃ§Ã£o das cores dos nÃ³s com base no grau (por exemplo, nÃ³s com alto grau sÃ£o azuis, nÃ³s com baixo grau com outras cores)
node_colors = list(node_degrees.values())

# DefiniÃ§Ã£o do tamanho dos nÃ³s com base no grau (por exemplo, nÃ³s com alto grau sÃ£o maiores)
node_sizes = [v * 100 for v in node_degrees.values()]

# Plota o grafo com nÃ³s coloridos e tamanhos personalizados
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Define o layout do grÃ¡fico

# Cria um objeto de mapa de cores para associar valores com cores
mapper = ScalarMappable(cmap=plt.cm.Blues)
mapper.set_array(node_colors)

# Desenhar os nÃ³s usando as cores e tamanhos personalizados
nodes = nx.draw_networkx_nodes(G, pos, node_color=node_colors, node_size=node_sizes, cmap=plt.cm.Blues)
edges = nx.draw_networkx_edges(G, pos, width=1.0, alpha=0.5)
nx.draw_networkx_labels(G, pos, font_size=10, font_weight='bold')

# Adicionando a barra de cores
plt.colorbar(mapper, label='Grau dos NÃ³s')

# Exibindo o grÃ¡fico
plt.title("Grafo Karate Club com Cores e Tamanhos Personalizados")
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt
# ParÃ¢metros da rede
N = 300  # NÃºmero de nÃ³s
L = 900  # NÃºmero de arestas
num_simulacoes = 1000  # NÃºmero de simulaÃ§Ãµes para obter a mÃ©dia

# Inicializa listas para armazenar resultados das simulaÃ§Ãµes
graus_medios = []
arestas_medias = []

#Calcular  A Probabilidade P
P = L/(N * (N -1)/2)

#Calcular
L = (N*(N - 1)* P)/2

# Realiza as simulaÃ§Ãµes
for _ in range(num_simulacoes):
    # Gera a rede ER
    G = nx.erdos_renyi_graph(N, p=0.02)  # Lembre-se de que p = L / (N * (N - 1) / 2)
    
    # Calcula o grau mÃ©dio e o nÃºmero de arestas
    grau_medio = sum(dict(G.degree()).values()) / N
    arestas_media = G.number_of_edges()
    
    # Armazena os resultados
    graus_medios.append(grau_medio)
    arestas_medias.append(arestas_media)

# Calcula a mÃ©dia dos resultados
grau_medio_medio = sum(graus_medios) / num_simulacoes
arestas_media_media = sum(arestas_medias) / num_simulacoes

# Imprime os resultados
print("Grau MÃ©dio Esperado:", 2 * L / N)
print("Grau MÃ©dio Simulado:", grau_medio_medio)
print("NÃºmero de Arestas Esperado:", L)
print("NÃºmero de Arestas Simulado:", arestas_media_media)

# Plota a rede
plt.figure(figsize=(8, 6))
ax = plt.gca()  # ObtÃ©m o objeto de eixo atual
nx.draw(G, with_labels=True, node_color='skyblue', node_size=100, font_size=10, font_weight='bold', edge_color='gray', ax=ax)
plt.title("Exemplo de Rede ER Gerada")
plt.show()
```

Para gerar uma rede ER correspondente com N=3000 nÃ³s e L=88.234.901 arestas, 
utilizando o modelo G(N,L). A probabilidade de conexÃ£o p Ã© dada por:
p = (2*L) / (N*(N-1))
  â 0.0000266

Asseguir vamos gerar a rede e comparar as distribuiÃ§Ãµes de grau com a rede original:

```python
import urllib.request
import matplotlib.pyplot as plt
import math

# Baixando a rede do Facebook do site SNAP
url = "https://snap.stanford.edu/data/facebook_combined.txt.gz"
filename = "facebook_combined.txt.gz"
urllib.request.urlretrieve(url, filename)

# Carregando a rede original do Facebook
G_orig = nx.read_edgelist(filename, nodetype=int)

# Gerando a rede ER correspondente
N = 3000

L = G_orig.number_of_edges()
p = (2*L) / (N*(N-1))
G_ER = nx.erdos_renyi_graph(N, p)

# Grau mÃ¡ximo e mÃ­nimo esperados
k_max = int((N-1)*p + 3*math.sqrt((N-1)*p*(1-p)))
k_min = int((N-1)*p - 3*math.sqrt((N-1)*p*(1-p)))

# Comparando as distribuiÃ§Ãµes de grau das duas redes
fig, ax = plt.subplots(1,2, figsize=(10,4))

ax[0].hist(dict(G_orig.degree()).values(), bins=range(100), alpha=0.5)
ax[0].set_title("DistribuiÃ§Ã£o de grau da rede original")
ax[0].set_xlabel("Grau")
ax[0].set_ylabel("FrequÃªncia")

ax[1].hist(dict(G_ER.degree()).values(), bins=range(100), alpha=0.5)
ax[1].set_title("DistribuiÃ§Ã£o de grau da rede ER correspondente")
ax[1].set_xlabel("Grau")
ax[1].set_ylabel("FrequÃªncia")

plt.show()


# Grau mÃ¡ximo e mÃ­nimo da rede original do Facebook
print("Rede original do Facebook:")
print(f"Grau mÃ¡ximo: {max(dict(G_orig.degree()).values())}")
print(f"Grau mÃ­nimo: {min(dict(G_orig.degree()).values())}")
print()

# Grau mÃ¡ximo e mÃ­nimo da rede ER correspondente
print("Rede ER correspondente:")
print(f"Grau mÃ¡ximo: {max(dict(G_ER.degree()).values())}")
print(f"Grau mÃ­nimo: {min(dict(G_ER.degree()).values())}")
```
